```python
import re

data_text = """
Посетите наши страницы:
Email: info@example.com (главный), support-team@corp.net (поддержка)
Дата: 25-05-2023, 01/01/2024, 12.02.2025.
Коды товаров: ID_PROD-1001, ITEM-20-A, ID-300-B.
Суммы: $150.50, 20 EUR, 500 RUB, £75.
"""

# Задание 1 — email
emails = re.findall(r'[\w\.\-]+@[\w\-]+\.[a-zA-Z]{2,3}', data_text)
print("Email-адреса:", emails)

# Задание 2 — коды товаров
codes = re.findall(r'ID(?:_PROD)?-(\d+)', data_text)
print("Цифровые части кодов:", codes)

# Задание 3 — замена формата даты
result = re.sub(r'(\d{2})\.(\d{2})\.(\d{4})', r'\3-\2-\1', data_text)
print("После замены дат:", result)

# Задание 4 — суммы в EUR и RUB
amounts = re.findall(r'\d+\s(?:EUR|RUB)', data_text)
print("Суммы в EUR/RUB:", amounts)
```
